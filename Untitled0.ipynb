```python
with open('A_Z Handwritten Data.csv', 'r') as csv_file:
    result = csv.reader(csv_file)
    rows = []

    for row in result:
        rows.append(row)
```

```python
train_data = [] 
train_label = [] 

for letter in rows:
    if True :
        x = np.array([int(j) for j in letter[1:]])
        x = x.reshape(28, 28)
        train_data.append(x)
        train_label.append(int(letter[0]))
    else:
        break
del rows[:]
```

```python
from sklearn.utils import shuffle
```

```python
train_data,train_label = shuffle(train_data,train_label)
```

```python
len = len(train_data)
train_x = train_data[:50000]
train_y = train_label[:50000]

val_x = train_data[50000:53000]
val_y = train_label[50000:53000]

test_x = train_data[53000:]
test_y = train_label[53000:]
del train_data[:]
del train_label[:]
```

```python
train_x = np.array(train_x).astype(np.float)/255
val_x = np.array(val_x).astype(np.float)/255
test_x  = np.array(test_x ).astype(np.float)/255
```

```python
train_x = train_x.reshape(-1, 28, 28, 1)
val_x = val_x.reshape(-1, 28, 28, 1)
test_x = test_x.reshape(-1, 28, 28, 1)
```

```python
train_x[6]
```

```python
from sklearn.preprocessing import LabelBinarizer
```

```python
lb = LabelBinarizer()
```

```python
train_y = lb.fit_transform(train_y)
```

```python
val_y = lb.fit_transform(val_y)
```

```python
#test_y = lb.fit_transform(test_y)
```

```python
train_y[0]
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.layers import AveragePooling2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import Activation
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Input
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import concatenate
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import SGD
```

```python
model = Sequential()
```

```python
width = 28
height = 28
classes = 26
```

```python
shape = (width, height, 1)
```

```python
model.add(Conv2D(32, (3, 3), padding="same", input_shape=shape))
```

```python
model.add(Activation("relu"))
```

```python
model.add(BatchNormalization())
```

```python
model.add(Conv2D(32, (3, 3), padding="same"))
```

```python
model.add(Activation("relu"))
```

```python
model.add(BatchNormalization())
```

```python
model.add(MaxPooling2D(pool_size=(2,2)))
```

```python
model.add(Conv2D(64, (3, 3), padding="same"))
```

```python
model.add(Activation("relu"))
```

```python
model.add(BatchNormalization())
```

```python
model.add(Conv2D(64, (3, 3), padding="same"))
```

```python
model.add(Activation("relu"))
```

```python
model.add(BatchNormalization())
```

```python
model.add(MaxPooling2D(pool_size=(2,2)))
```

```python
model.add(Flatten())
```

```python
model.add(Dense(512))
```

```python
model.add(Activation("relu"))
```

```python
model.add(BatchNormalization())
```

```python
model.add(Dense(classes))
```

```python
model.add(Activation("softmax"))
```

```python
model.summary()
```

```python
dataAugmentaion = ImageDataGenerator(rotation_range = 30, zoom_range = 0.20, 
fill_mode = "nearest", shear_range = 0.20, horizontal_flip = True, 
width_shift_range = 0.1, height_shift_range = 0.1)
```

```python
dataAugmentaion.flow(train_x, train_y, batch_size = 32)
```

```python
learning_rate = 0.01
```

```python
epochs = 10
# epoch
# Steps
# 
batch_size = 64
```

```python
opt = SGD(learning_rate=learning_rate, momentum=0.9)
```

```python
model.compile(optimizer=opt, loss="categorical_crossentropy", metrics=["accuracy"])
```

```python
print("Start training")
H = model.fit_generator(dataAugmentaion.flow(train_x, train_y, batch_size=batch_size), validation_data=(val_x, val_y), steps_per_epoch=train_x.shape[0]//batch_size, epochs=epochs, verbose=1)
```

```python
# dự đoán với tập dữ liệu test
test_logits = model.predict(test_x[1:2])
#lấy phần tử có giá trị lớn nhất 
test_logits = np.argmax(test_logits, axis=-1)
```
